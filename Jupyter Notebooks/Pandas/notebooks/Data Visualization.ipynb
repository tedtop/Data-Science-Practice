## How has the injury rate changed over time?

To answer this question, we'll focus on the `INJURIES_TOTAL` column, which records the total number of individuals who sustained fatal, incapacitating, non-incapacitating, or possible injuries as determined by the reporting officer.

First, we'll create a new column named `INJURIES` that will be set to 1 if there was one or more injuries in a crash, and 0 if there were none.

```python
# INJURIES = 1 if the crash involves injuries
df['INJURIES'] = (df.INJURIES_TOTAL>=1).astype(int) 
df.INJURIES
```

The `.astype(int)` method converts True to 1 and False to 0. 
While this step isn't strictly necessary, as Python treats True as 1 and False as 0 by default, it ensures consistency in the data format.

Let's plot the weekly number of crashes involving one or more injuries, as well as the weekly number of crashes with no injuries.

we'll start by grouping the data by the `INJURIES` column and then resample it on a weekly basis. After that, we'll select any column and count the number of crashes in each group.

This will produce a DataFrame with a multi-index, where `INJURIES` and `CRASH_DATE` are the two index levels. Finally, we'll "unstack" the `INJURIES` level (level 0) to separate the data into two columns: one for crashes with injuries and one for those without.

```python
df.groupby('INJURIES').resample('W').CRASH_RECORD_ID.count()
```

If you were to try resampling first and then grouping, you would encounter the error 'DatetimeIndexResampler' object has no attribute 'groupby' because resample() returns a resampler object that doesnâ€™t support groupby directly.

```python
# Plot the weekly crash counts
df.groupby('INJURIES').resample('W').CRASH_RECORD_ID.count().unstack(level=0).plot(figsize=(15,5))
```

Finally, let's calculate the percentage of crashes that involve one or more injuries. We'll create two Series: one for the total number of crashes and another for crashes that involve injuries.

```python
# number of crashes that involve injuries
n_injuries = df[df.INJURIES==1].resample('W').CRASH_RECORD_ID.count()
n_injuries
```

```python
# total number of crashes
n_total = df.resample('W').CRASH_RECORD_ID.count()
n_total
```

```python
# percentage of crashes that invole injuries 
pct_weekly_injuries = 100*n_injuries/n_total
pct_weekly_injuries.plot(figsize=(15,5))
```

## How does the injury rate vary throughout the week?

Let's create a table where the index represents the days of the week (which we can extract from the DataFrame index), and the columns are labeled 0 (no injuries) and 1 (injuries).

```python
table = df.pivot_table(values='CRASH_RECORD_ID', index=df.index.dayofweek, columns='INJURIES',aggfunc='count')
table
```

Let's add a column for the total number of crashes per day (by summing across the columns) and another column for the percentage of crashes involving injuries.

```python
table['total'] = table.sum(axis=1)
table['pct_injuries'] = 100*table[1]/table['total']
table.index = ['Mon','Tue','Wed','Thu','Fri','Sat','Sun']
table
```

Finally, let's visualize the percentages using a bar plot.

```python
table['pct_injuries'].plot.bar()
```

```python
# alternatively
table['pct_injuries'].plot(kind='bar')
```

Sundays appear to be the most dangerous day of the week.

## Advanced plot

Pandas is built on top of Matplotlib, so we can manually enhance our plots using Matplotlibâ€™s features.

Let's calculate the weekly number of crashes involving injuries and those without injuries.

```python
# weekly number of crashes by injury status (0=no, 1=yes)
weekly = df.groupby('INJURIES').resample('W').CRASH_RECORD_ID.count().unstack(level=0)
weekly
```

Previously, we plotted this data using the `.plot()` method.

```python
weekly.plot(figsize=(15,7))
```

But now, we can use Matplotlib directly to customize our plot further.

In Matplotlib, a figure is the overall container that holds everything you see on the plot, including one or more plots, titles, and labels. An axis refers to the specific area where the data is plotted, including the x-axis and y-axis. You can think of the figure as the canvas, and the axis as the part of the canvas where the actual drawing happens.

```python
# Create a figure and axis with a specified size
fig, ax = plt.subplots(figsize=(15,5))

# Plot the weekly DataFrame onto the created axis
weekly.plot(ax=ax)

# Set the x-axis label with a specific font size
ax.set_xlabel('Date', fontsize=15)

# Set the y-axis label with a specific font size
ax.set_ylabel('Weekly Number of Crashes', fontsize=15)

# Add a title to the plot with a specific font size
ax.set_title('Number of Traffic Crashes per Week', fontsize=20)

# Customize the legend with labels, font size, location, and title
ax.legend(labels=['No Injuries', 'With Injuries'], fontsize=15, loc='upper left', title='Injuries?', title_fontsize=18)

# Add grid lines to the plot for better readability
ax.grid(True)
```

One drawback of the previous plot is that the two lines are on very different scales, making it difficult to visually compare them. 
To overcome this, we can create two separate plots, displayed side by side, each showing one of the line plots. 
We can achieve this using `plt.subplots`, which allows us to create multiple subplots in a single figure, providing a clearer comparison.

In the code below, `fig, ax = plt.subplots(1, 2, figsize=(15, 5))` will create a figure with two subplots arranged in a single row (1 row, 2 columns), with a total figure size of 15x5 inches. 
The variable `ax` will hold an array of the two subplot axes, allowing us to plot each line on a separate axis.

```python
# Create a figure with two subplots side by side (1 row, 2 columns) and set the figure size
fig, ax = plt.subplots(1, 2, figsize=(15, 5))

# Plot the weekly data for no injuries on the first subplot (left)
weekly[0].plot(ax=ax[0])

# Plot the weekly data for crashes with injuries on the second subplot (right), using a red color
weekly[1].plot(ax=ax[1], color='red')

# Set the x-axis labels for both subplots with a specific font size
ax[0].set_xlabel('Date', fontsize=15)
ax[1].set_xlabel('Date', fontsize=15)

# Set the titles for each subplot to indicate the injury status
ax[0].set_title('Injuries = 0')
ax[1].set_title('Injuries = 1')

# Add a common title for the entire figure
fig.suptitle('Number of Traffic Crashes per Week', fontsize=20)

# Add grid lines to both subplots for better readability
ax[0].grid(True)
ax[1].grid(True)
```

An alternative to the above plot is to create a twin axis that shares the same x-axis but has separate y-axes for each dataset. 
This is helpful when the two datasets (e.g., crashes with and without injuries) have different scales, allowing for easier comparison.
The code uses `axis.twinx()` to create this twin axis, so both datasets can be plotted together while keeping separate y-axes. This lets you see both trends on the same timeline.

```python
# Create a figure and primary axis
fig, ax = plt.subplots(figsize=(15, 5))

# Create a secondary axis that shares the same x-axis
ax2 = ax.twinx()

# Plot the weekly data for crashes with no injuries on the primary axis
weekly[0].plot(ax=ax)

# Plot the weekly data for crashes with injuries on the secondary axis, using a different color
weekly[1].plot(ax=ax2, color='C1')

# Set the label for the x-axis with a specific font size
ax.set_xlabel('Date', fontsize=15)

# Set the y-axis label for the primary axis (no injuries)
ax.set_ylabel('Injuries = 0', fontsize=15)

# Set the y-axis label for the secondary axis (injuries)
ax2.set_ylabel('Injuries = 1', fontsize=15)

# Add a title to the plot with a specific font size
ax.set_title('Number of Traffic Crashes per Week', fontsize=20)

# Add a legend to the figure, with customized labels, font size, location, and title
fig.legend(labels=['No Injuries', 'Injuries'], fontsize=15, loc='upper left', title='Injuries?', title_fontsize=18)

# Add grid lines to the primary axis for better readability
ax.grid(True)
```
