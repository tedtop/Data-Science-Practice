```python
import pandas as pd
df = pd.read_csv('../../../Data/traffic_crashes_Chicago.csv', index_col='CRASH_DATE')
df
```

```python
df.sort_index(inplace=True, ascending=False)
df.head()
```

```python
df.columns
```

```python
# filter out incomplete data. become complete Sep 2017
# filter out up to 2018
df.index = pd.to_datetime(df.index)
df = df[df.index.year >= 2018]
df
```

## How has the weekly number of crashes changed over time?

```python
df.resample('W').CRASH_RECORD_ID.count()
```

```python
df.resample('W').CRASH_RECORD_ID.count().plot(figsize=(15, 7))
```

```python
df.INJURIES_TOTAL.max()
```

```python
df['INJURIES'] = (df.INJURIES_TOTAL >= 1).astype(int)
df.INJURIES.sum()
```

```python
df.groupby('INJURIES').resample('W').CRASH_RECORD_ID.count().unstack(level=0).plot()
```

```python
# what percetage of crashes result in injuries
df.INJURIES.sum() / len(df)
```

```python
# compute the number of injuries per week
#df.INJURIES_TOTAL.resample('W').sum().plot()
n_inuries = df[df.INJURIES == 1].resample('W').CRASH_RECORD_ID.count()
n_inuries
```

```python
n_total = df.resample('W').CRASH_RECORD_ID.count()
n_total
```

```python
pct_weekly_injuries = n_inuries / n_total * 100
pct_weekly_injuries.plot()
```
