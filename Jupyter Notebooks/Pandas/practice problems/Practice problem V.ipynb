# Practice problem V

```python
# load the Chipotle dataset
path = 'https://raw.githubusercontent.com/um-perez-alvaro/Data-Science-Practice/master/Data/chipotleorders.csv'
df = pd.read_csv(path)
df
```

Sort the dataframe by the name of the item

```python
df.sort_values('item_name')
```

What was the most expensive item ordered.

```python
df.sort_values('item_price', ascending=False)
# the most expensive order was 15 orders of Chips and Fresh Tomato Salsa, which cost $44.25
```

```python
# however, the most expensive item was the Steak Salad Bowl, which cost $11.89
df[df['quantity'] == 1].sort_values('item_price', ascending=False)
```

How many times was a Veggie Salad Bowl ordered?

```python
len(df[df['item_name'] == 'Veggie Salad Bowl'])
```

How many times did someone order more than one Canned Soda?

```python
len(df[(df['item_name'] == 'Canned Soda') & (df['quantity'] > 1)])
```
