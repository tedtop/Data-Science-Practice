```python
data = pd.read_csv('universities_2016.csv',index_col='University_Name')
```

```python
data.dtypes
```

```python
# feature matrix
X = data.drop(['World_Rank','Country','Year','Total_Score'],axis=1)
```

```python
X.dtypes
```

```python
# missing values?
data.isna().sum()
```

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler

k = 5
pipe = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='mean')),
    ('scaler', MinMaxScaler()),
    ('cluster', KMeans(n_clusters=5))
])

pipe
```

```python
pipe.fit(X)
```

```python
labels = pipe['cluster'].labels_
centers = pipe['cluster'].cluster_centers_
```

```python
X.columns
```

```python
len(X.columns)
```

```python
for i in range(k):
    plt.plot(centers[i],'o--')
plt.xticks(ticks = range(9), labels=X.columns, rotation=90)
plt.show()
```
